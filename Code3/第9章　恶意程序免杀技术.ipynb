# 9.2　文件末尾追加随机内容

```python
import random

min_append_log2 = 5
max_append_log2 = 8

def __random_length():
return 2**random.randint(min_append_log2, max_append_log2)

upper = random.randrange(256)
L = __random_length()
bytez += bytes([random.randint(0, upper) for _ in range(L)])
bytez
```

# 9.3　追加导入表

```python
import json
COMMON_IMPORTS = json.load(open(r'gym_malware\small_dll_imports.json','r'))
COMMON_IMPORTS
```

```python
libname = random.choice(list(COMMON_IMPORTS.keys()))
funcname = random.choice(list(COMMON_IMPORTS[libname]))
lowerlibname = libname.lower()
libname,funcname,lowerlibname
```

```python
# 解析PE文件，获取当前的导入表，遍历导入表，如果随机选择的库不存在，就创建一个新库，如果导入函数也不存在，就增加一个新入口
binary = lief.parse(bytez)
print(binary.imports)
lib = None
for im in binary.imports:
if im.name.lower() == lowerlibname:
lib = im
break

if lib is None:
lib = binary.add_library(libname)
names = set([e.name for e in lib.entries])
if not funcname in names:
lib.add_entry(funcname)
lib
```

# 9.4　改变节名称

```python
COMMON_SECTION_NAMES = open('gym_malware\section_names.txt','r').read().rstrip().split('\n')
binary = lief.parse(bytez)
targeted_section = random.choice(binary.sections)
print("改变前:",targeted_section.name)
targeted_section.name = random.choice(COMMON_SECTION_NAMES)[:7]
print("改变后:",targeted_section.name)
```

# 9.5　增加节

```python
binary = lief.parse(bytez)
# 随机生成节名称的方法，在ASCII表中“.”～“z”之间随机生成6位字符
new_section = lief.PE.Section("".join(chr(random.randrange(ord('.'), ord('z'))) for _ in range(6)))
print(new_section)
```

```python
# 随机生成长度，并随机生成内容
upper = random.randrange(256)
L = __random_length()
print(upper, L)
new_section.content = [random.randint(0, upper) for _ in range(L)]
print(new_section)
```

```python
binary.add_section(new_section,
                    random.choice([
                    lief.PE.SECTION_TYPES.BSS,
                    lief.PE.SECTION_TYPES.DATA,
                    lief.PE.SECTION_TYPES.EXPORT,
                    lief.PE.SECTION_TYPES.IDATA,
                    lief.PE.SECTION_TYPES.RELOCATION,
                    lief.PE.SECTION_TYPES.RESOURCE,
                    lief.PE.SECTION_TYPES.TEXT,
                    lief.PE.SECTION_TYPES.TLS_,
                    lief.PE.SECTION_TYPES.UNKNOWN,]))
print(binary.sections)
```
